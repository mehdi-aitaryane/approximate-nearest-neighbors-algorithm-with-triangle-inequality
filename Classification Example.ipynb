```python
from neighbors import ANNeighborClassifier
from datasets import make_classification, make_blobs, make_circles
from splitters import splitXy
from plots import scatter2D, plot_db_2d
import numpy as np
import time
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=4, noise=0.0, balanced=True, random_state=54)
```

```python
title = "Make Classification Dataset"
xlabel = "Feature 1"
ylabel = "Feature 2"
scatter2D(title, xlabel, ylabel, X, y)
```

```python
X_train, X_test, y_train, y_test = splitXy(X, y, test_size=0.2, random_state = 111)
```

```python
start = time.time()
ann = ANNeighborClassifier()
ann.fit(X_train, y_train)
score = ann.score(X_test, y_test)
end = time.time()
print("Approximate Nearst Neighbor Classifier Accuracy is ", np.round(100 * score, 2), "%")
print('Time taken ', np.round(end-start, 2),' seconds')
```

```python
start = time.time()
title = "Decision Boundary"
xlabel = "Feature 1"
ylabel = "Feature 2"
plot_db_2d(ann, X_test, y_test, title, xlabel, ylabel, step=50)
end = time.time()
print('Time taken ', np.round(end-start, 2),' seconds')
```
